```python
import pandas as pd
import numpy as np

df = pd.read_csv("insurance_data.csv")
print(df.head())

print("Dataset Info:")
df.info()

print("Descriptive Statistics:")
print(df.describe())

df.fillna("Unknown", inplace=True)

df['TXN_DATE_TIME'] = pd.to_datetime(df['TXN_DATE_TIME'], errors='coerce')

df.drop_duplicates(subset=['TRANSACTION_ID'], inplace=True)

bins = [0, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
labels = ['Under 5k', '5k to 10k', '10k to 20k', '20k to 30k', '30k to 40k', '40k to 50k', 
          '50k to 60k', '60k to 70k', '70k to 80k', '80k to 90k', '90k to 100k']

df['Claim Amount Bins'] = pd.cut(df['CLAIM_AMOUNT'], bins=bins, labels=labels, include_lowest=True)

claim_distribution = df['Claim Amount Bins'].value_counts()
print('Claim distribution:'+ str(claim_distribution))

correlation = df[['CLAIM_AMOUNT', 'PREMIUM_AMOUNT']].corr()
correlation_value = correlation.loc['CLAIM_AMOUNT', 'PREMIUM_AMOUNT']
print('Correlation coefficient: ' + str(correlation_value))

df.to_csv("cleaned_insurance_data.csv", index=False)
```
